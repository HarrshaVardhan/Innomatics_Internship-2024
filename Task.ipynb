```python
import pandas as pd

# Assuming you have a DataFrame named 'ratings' with columns 'movieId', 'title', and 'rating'
# and a DataFrame named 'movies' with information about genres
# Load the data from your CSV file or any other source
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')  # Assuming you have a 'genres' column in this DataFrame

# Merge the ratings and movies DataFrames
merged_data = pd.merge(ratings, movies, on='movieId')

# Filter for Sci-Fi movies
sci_fi_movies = merged_data[merged_data['genres'].str.contains('Sci-Fi')]

# Count the number of ratings for each Sci-Fi movie
ratings_count_sci_fi = sci_fi_movies.groupby('title')['rating'].count()

# Sort the Sci-Fi movies based on the number of ratings in descending order
sorted_sci_fi_movies = ratings_count_sci_fi.sort_values(ascending=False)

# Identify the third most popular Sci-Fi movie
third_most_popular_sci_fi_movie = sorted_sci_fi_movies.index[2]

print("Third most popular Sci-Fi movie based on number of user ratings:")
print(third_most_popular_sci_fi_movie)
```

```python
import pandas as pd

def find_highest_id(file_path):
    try:
        # Read the CSV file
        df = pd.read_csv("links.csv")
        
        # Ensure the 'id' column exists
        if 'imdbId' in df.columns:
            # Find and return the highest ID
            return df['imdbId'].max()
        else:
            print("The 'id' column does not exist in the file.")
            return None
    except FileNotFoundError:
        print(f"The file {file_path} was not found.")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Example usage
file_path = 'links.csv'
highest_id = find_highest_id(file_path)

if highest_id is not None:
    print(f"The highest ID in {file_path} is: {highest_id}")
```

```python
import pandas as pd

def find_highest_rated_sci_fi_movie(movies_file_path, ratings_file_path):
    try:
        # Read the CSV files
        movies_df = pd.read_csv("movies.csv")
        ratings_df = pd.read_csv("ratings.csv")
        
        # Filter Sci-Fi movies
        sci_fi_movies_df = movies_df[movies_df['genres'].str.contains('Sci-Fi')]
        
        # Merge with ratings
        merged_df = pd.merge(sci_fi_movies_df, ratings_df, on='movieId')
        
        # Calculate average rating for each movie
        avg_ratings_df = merged_df.groupby('movieId')['rating'].mean().reset_index()
        
        # Find the movieId with the highest average rating
        highest_rated_movie_id = avg_ratings_df.loc[avg_ratings_df['rating'].idxmax()]['movieId']
        
        return highest_rated_movie_id
    except FileNotFoundError:
        print(f"One of the files was not found.")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Example usage
movies_file_path = 'movies.csv'
ratings_file_path = 'ratings.csv'
highest_rated_movie_id = find_highest_rated_sci_fi_movie(movies_file_path, ratings_file_path)

if highest_rated_movie_id is not None:
    print(f"The movieId of the highest rated Sci-Fi movie is: {highest_rated_movie_id}")
```
